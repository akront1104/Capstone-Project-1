Firefighter Fatalities in the United States
2000 – 2015

The objective of the U.S. Fire Administration (USFA) study “Firefighter Fatalities in the United States” is to identify and analyze all on-duty firefighter fatalities to increase understanding of their causes and how they can be prevented. The study is intended to help identify approaches that could reduce the number of deaths in future years.

On Dec. 15, 2003, the president of the U.S. signed the Hometown Heroes Survivors Benefit Act of 2003 into
law. The law presumes that a heart attack or stroke is in the line of duty if the firefighter was engaged in nonroutine, stressful or strenuous physical activity while on-duty, and the firefighter becomes ill while on-duty, or within 24 hours after engaging in such activity. My analysis aims to go further into Career and Volunteer Firefighter fatalities caused by heart attacks.

```python
#1) Creating function to count cause of death by type
def count_cause_by_type (df, nature_of_death, age_criteria1, age_criteria2=1):
    """This function will return the number of fatalities by a specific Cause Of Death type that fit a certain age criteria"""
    
    # Filters age in DataFrame based on age criteria
    filter_age = df[np.logical_and(df['Age'] >= age_criteria2, df['Age'] < age_criteria1)]
    
    # Filters age DataFrame by 'Nature Of Death' and counts
    causes_by_nature_count = filter_age['Nature Of Death'].value_counts()[nature_of_death]
    
    #returns number of fatalities of specific nature of death that fall in the age criterias
    return causes_by_nature_count
```

```python
#2) Calling function "count_cause_by_type" using stress_df_vol DataFrame

print(" Number of fatalities caused by heart attacks among Volunteer Firefigthers, by age group - US (2000 -2015)\n")

#a) Number of heart attacks among volunteer firefighters, under 25 years 
vol_heart_attack_age_under_25 = count_cause_by_type(stress_df_vol, 'Heart Attack', 25)
print("< 25: " + str(vol_heart_attack_age_under_25))

#b) Number of heart attacks among volunteer firefighters, aged 25 - 34 
vol_heart_attack_age_25_34 = count_cause_by_type(stress_df_vol, 'Heart Attack',35, age_criteria2=25)
print("25 - 34: " + str(vol_heart_attack_age_25_34))

#c) Number of heart attacks among volunteer firefighters, aged 35 - 44
vol_heart_attack_age_35_44 = count_cause_by_type(stress_df_vol, 'Heart Attack',45, age_criteria2=35)
print("35 - 44: " + str(vol_heart_attack_age_35_44))

#d) Number of heart attacks among volunteer firefighters, aged 45 - 54 
vol_heart_attack_age_45_54 = count_cause_by_type(stress_df_vol, 'Heart Attack',55, age_criteria2=45)
print("45 - 54: " + str(vol_heart_attack_age_45_54))

#e) Number of heart attacks among volunteer firefighters, aged 55 - 65 
vol_heart_attack_age_55_65 = count_cause_by_type(stress_df_vol, 'Heart Attack',66, age_criteria2=55)
print("55 - 65: " + str(vol_heart_attack_age_55_65))

#f) Number of heart attacks among volunteer firefighters, over 65 years
vol_heart_attack_age_over_65 = count_cause_by_type(stress_df_vol, 'Heart Attack',100, age_criteria2=66)
print("> 65: " + str(vol_heart_attack_age_over_65) +"\n")

#g) Total Volunteer Firefighter Heart Attack Fatalities
total_vol_heart_attack = vol_heart_attack_age_under_25 + vol_heart_attack_age_25_34 + vol_heart_attack_age_35_44 + vol_heart_attack_age_45_54 + vol_heart_attack_age_55_65 + vol_heart_attack_age_over_65  
print('Total: ' + str(total_vol_heart_attack) + "\n\n")


#3) Calling function "count_cause_by_type" using stress_df_car DataFrame

print(" Number of fatalities caused by heart attacks among Career Firefigthers, by age group - US (2000 -2015)\n")

#a) Number of heart attacks among career firefighters, under 25 years 
car_heart_attack_age_under_25 = count_cause_by_type(stress_df_car, 'Heart Attack', 25)
print("< 25: " , car_heart_attack_age_under_25)

#b) Number of heart attacks among career firefighters, aged 25 - 34 
car_heart_attack_age_25_34 = count_cause_by_type(stress_df_car, 'Heart Attack',35, age_criteria2=25)
print("25 - 34: " + str(car_heart_attack_age_25_34))

#c) Number of heart attacks among career firefighters, aged 35 - 44 
car_heart_attack_age_35_44 = count_cause_by_type(stress_df_car, 'Heart Attack',45, age_criteria2=35)
print("35 - 44: " + str(car_heart_attack_age_35_44))

#d) Number of heart attacks among career firefighters, aged 45 - 54 
car_heart_attack_age_45_54 = count_cause_by_type(stress_df_car, 'Heart Attack',55, age_criteria2=45)
print("45 - 54: " + str(car_heart_attack_age_45_54))

#e) Number of heart attacks among career firefighters, aged 55 - 65 
car_heart_attack_age_55_65 = count_cause_by_type(stress_df_car, 'Heart Attack',66, age_criteria2=55)
print("55 - 65: " + str(car_heart_attack_age_55_65))

#f) Number of heart attacks among career firefighters, over 65 years
car_heart_attack_age_over_65 = count_cause_by_type(stress_df_car, 'Heart Attack',100, age_criteria2=66)
print("> 65: " + str(car_heart_attack_age_over_65) +"\n")

# Total Career Firefighter Heart Attack Fatalities
total_car_heart_attack = car_heart_attack_age_under_25 + car_heart_attack_age_25_34 + car_heart_attack_age_35_44 + car_heart_attack_age_45_54 + car_heart_attack_age_55_65 + car_heart_attack_age_over_65
print('Total: ' + str(total_car_heart_attack))
```
